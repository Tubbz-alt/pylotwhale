```python
import pylab as pl
import numpy as np
import sys
import os
import scikits.audiolab as al

import plSpecgrams as pspec
```

```python
def time_line(wf_length, sampling_rate):
    vec = np.arange(0.0, 1.0*wf_length/sampling_rate, 1.0/sampling_rate)
    if(len(vec) == wf_length):
        return vec
    if(len(vec) == wf_length+1):
        return np.delete(vec, len(vec)-1)   
```

### Read WAV

```python
wavF= '/home/florencia/whales/reports/NLP/manuscript/supplement/wavs/secsJ/WAV_2009-07-13-0050_008-48kHz.wav'
waveform, samplingR, inEnc, = al.wavread(wavF)
print "lenght:", len(waveform), ", sampling freq:", samplingR, ", encoding:", inEnc
```

```python
#play
al.play(waveform, fs=samplingR)
```

```python
# plot waveform
WFlen = len(waveform)
pl.figure(figsize=(16,3))
pl.subplot(111)
pl.plot( time_line(len(waveform), samplingR), waveform)
```

```python
# cutting parameters
tc_0 = 36#14.02*60
tc_f = 90
idLab = '_%d-%dsec.wav'%(tc_0, tc_f) #out file name label
Tf = time_line(len(waveform), samplingR)[-1] #time
ix_0 = int(tc_0*WFlen/Tf) # begining index
ix_f = int(tc_f*WFlen/Tf) # ending index
cuttedWF = waveform[ix_0:ix_f] # cutted waveform

print "cutting indexes", ix_0, ix_f

print "waveform length:", WFlen, "min:", np.min(waveform), ",max:", np.max(waveform)#, ", time: (%.1f, %.2f)"%(time[0],time[-1])
pl.figure(figsize=(16,3))
pl.subplot(111)
pl.plot( cuttedWF)
#pl.xlim(13,16)

outWavFN = wavF.split('.wav')[0]+idLab
print outWavFN
```

```python
# write

format = al.Format('wav')

new_file = al.Sndfile(outWavFN, 'w', format, 1, samplingR) # declare a writing audio file
new_file.write_frames(cuttedWF)
new_file.close()

print "find me here:", os.path.abspath(outWavFN)
```

```python
pspec.saveSpecgram(outWavFN, 10)
```
