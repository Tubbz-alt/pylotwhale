#Filter in the time domain

###Filter functions

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

##Power spectral functions

```python
for i in range(4,5):
    print(2**i)
    
textureWS = 0.1
step = int(fs*textureWS)    
#n=int(np.log(step)/)
print("n", n, "\nws(samples)", 2**n)
```

```python
def bandEnergy(y, fs):
    '''
    returns the power spectral energy of y
    '''
    f, Pxx_den = welch(y, fs, nperseg=256)
    bandPSD = np.sum(Pxx_den**2)
    return(bandPSD)

def tsBandenergy(y, fs, textureWS=0.1, textureWSsamps=0, overlap=0):
    '''
    returns  a time series with the spectral energy every textureWS (seconds)
    overlap [0,1)
    < signal
    < sampling rate
    < textureWS in seconds
    < textureWSsamps in number of samples, 
    -->
    > time series of the powerspectral density
    > time interval of the texture window
    '''
    step = 2**int(np.log(fs*textureWS)/np.log(2))
    overl = int(step*overlap)
    PSts = []
    ix0 = 0
    ix = step
    print("step:", step, "overlap", overl, "len signal", len(y), "step (s)", 1.0*step/fs)
    while ix<len(y):
        yi = y[ix0:ix]
        #print("interval: ", ix0, ix, len(yi))
        ix0 = ix-overl
        ix = ix0+step
        #print(ix0)

        PSts.append(bandEnergy(yi, fs))
    print(len(PSts))
    return(PSts, 1.0*step/fs)


def bandDecomposition(x, fs, intervals=None, filFun = sT.butter_bandpass_filter, order=3, Nbins=4):
    '''
    band filter
    
    < x, input signal
    < fs, sampling rate
    < intervals, list of tuples with intervals of the bands.
                if None, then (Nbins) are defined logarithmic sizes
    < order, order of the butter filter
    < bins, used when the intervals are defined (default log)
    -->
    > yLi, dictionary of the dand decomposition of x 
    
    '''   
    
    if intervals == None:
        li = np.array([int(item) for item in np.logspace(np.log(1000), np.log(fs/2), num=Nbins, base=np.e)])
        intervals = []
        i0 = 100
        for item in li:
            intervals.append((i0, item))
            i0 = item
    assert type(intervals) == list, '! intervals must be a list'
    
    yLi = {}
    for (lcut, ucut) in intervals:
        print("input", lcut, ucut, np.shape(x))
        yLi[(lcut, ucut)] = sT.butter_bandpass_filter(x, lcut, ucut, fs, order=order)
            
    return(yLi)    
#print("original waveform", bandEnergy(x, fs), "\nfiltered", bandEnergy(y, fs))
```
