```python
def show_Specgram(speech):
    '''
    Reads .wav file from STDIN and plots the spectrogram
    '''
    sound = al.sndfile(speech,'read')
    # Reads wav file with audiolab
    sound_info = sound.read_frames(sound.get_nframes())
    # Extracts feature info from sound file with scipy module
    # sound_info will have 1 column for monaural data, 2 for stereo etc.
    figure( )
    subplots_adjust(hspace=.5)
    # get the sample rate, this will ensure the proper axes in the spectorgram.
    Fs = sound.get_samplerate()
    window = mlab.window_hanning
    nowindow = window_none
    #Generates soectrogram with matplotlib specgram

    print "sound info", sound_info
    
    #check the size of the array
    
    figure(figsize=(18,6))
    subplot(131)
    spectrogram = pl.specgram(np.atleast_2d(sound_info)[0,:],Fs=Fs,window=nowindow)
    spectrogram=specgram
    title('Spectrogram:  nowindow')
    subplot(132)
    spectrogram = pl.specgram(np.atleast_2d(sound_info)[0,:],Fs=Fs,window=window)
    spectrogram=specgram
    title('Spectrogram:  Hanning')
    subplot(133)
    spectrogram = pl.specgram(np.atleast_2d(sound_info)[0,:],Fs=Fs,window=window,noverlap=100)
    spectrogram=specgram
    title('Spectrogram:  Hanning, overlap 100')
    show()
    sound.close()
    return spectrogram


#file = sys.argv[1]

#show_Specgram(file)
```
