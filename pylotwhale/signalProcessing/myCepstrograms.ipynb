```python
baseN = os.path.splitext(fileN)[0]

print "base name:", baseN, "\nbase name", os.path.basename(fileN).split('.')[0], "\ndir", os.path.dirname(fileN)
```

### Settings

```python
pw = 9
NFFT = 2**pw
pad_to = NFFT
over = 0.9

noverlap = int(NFFT*over) #NFFT-1#
numFreqs = pad_to //2
winN = 'hamming'
windowVals = sig.get_window(winN, NFFT)#np.hamming(N)
scaling_factor = 2
step = NFFT - noverlap
ind = np.arange(0, len(x) - NFFT + 1, step)
qf = 1.0*NFFT/(2.0*sRate)

numCeps = numFreqs//2
windowC = sig.get_window('hamming',numFreqs)

print "singal", len(waveform), "NFFT", NFFT, "overlap", noverlap, "num freqs", numFreqs, "Nceps", numCeps, "final quefreq", qf
```

### compute cepstrogram

```python
n = len(ind)

#Pxy = np.zeros((numFreqs, n), np.complex_)
Cxy = np.zeros((numCeps, n), np.complex_)

for i in range(n):
        thisX = x[ind[i]:ind[i]+NFFT]
        thisX = windowVals * thisX#* detrend(thisX) #flo: window the singal, detrend desplaces the signal to cero mean
        #thisX = windowVals * thisX #+flo: window the singal
        #fx = np.fft.fft(thisX, n=pad_to) -flo
        fx = ( np.fft.fft(thisX, n=pad_to)) #+flo: fourier transform
        Px = np.conjugate(fx[:numFreqs]) * fx[:numFreqs] #+flo: abs
        Px /= (np.abs(windowVals)**2).sum() #+flo: scale
        Px[1:-1] *= scaling_factor #+flo
        Cx = ( np.fft.fft(windowC*np.log(Px))) #+flo: fourier transform
        Cxy[:,i] = np.conjugate(Cx[:numCeps]) * Cx[:numCeps] #flo: power scpectrum: 1) tae

#Cxy /= (np.abs(windowC)**2).sum() #flo: divide each element by the power of the window
#Cxy[1:-1] *= scaling_factor
        
```
