```python
import numpy as np
import matplotlib.pyplot as plt
```

Measuring the periodicity of a signal with **autocorrelation.**

```python
sigLen = 2000
x = np.arange(0, sigLen, 1)  # time
# Our  signal
y = np.zeros(len(x)) 
deltaPeriod = 500
y[np.arange(100, sigLen, deltaPeriod)]=1 # delta functions every 
y[np.arange(0, sigLen, 200)]=1 # delta functions every 
```

let's visualize the signal and it's fourier transform

```python
fig, ax =  plt.subplots(1, 2, figsize = (15,4))
ax[0].plot(x,y)
ax[1].plot(np.fft.fft(y))
```

```python
#plt.plot(np(y))
plt.plot(np.correlate( y, y, mode = 'full'))
#plt.xlim(0, 20000)
```
