```python
import numpy as np
from numpy.fft import fft
import sys
import scipy as sp
import pylab as pl
import matplotlib
```

### Functions

```python
def time_line(wf_length, sampling_rate):
    vec = np.arange(0.0, 1.0*wf_length/sampling_rate, 1.0/sampling_rate)
    if(len(vec) == wf_length):
        return vec
    if(len(vec) == wf_length+1):
        return np.delete(vec, len(vec)-1)   
```

```python
def myStep(start, pointsPstep, mylenght = 10000000):
    pStep  = 1.0/pointsPstep
    stepVec = np.round((np.arange(mylenght)*pStep)+1.0)*start
    print len(stepVec)
    pl.plot(stepVec)
    return stepVec
```

```python
def myHarmonic(f0, numHar, t, phase = 0.0):
    '''
    this function returns a oscilating dital signal (of t) with fundamental frequency f0, and numHar harmonics
    '''
    signal = 0.0
    for n in np.arange(1,numHar+1):
        signal = signal + np.sin(2.0*pi*f0*n*t + phase)
        print n, numHar
    return signal/(numHar+1) # we normalize the function, so that we can later save it as a wav
```
