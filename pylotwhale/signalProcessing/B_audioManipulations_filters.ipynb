```python
import numpy as np
import scipy.signal as sig
import scipy.io.wavfile as wavfile
import pdb
#from numpy import 

import matplotlib.pyplot as plt
import pylotwhale.signalProcessing.signalTools_beta as sT

%matplotlib inline
```

```python
def plotSpectrum(y, Fs):
    Y = np.fft.rfft(y)
    x = np.linspace(0, Fs/2, len(Y))
    plt.plot(x, Y)
```

```python
audioFile = "/home/florencia/whales/data/Vocal-repertoire-catalogue-Pilot-whales-Norway/flo/wavs/callClassifier_prototypes_noiseReduced/NPW-093ii-B-B-100703_f108-1_00_01_07.wav"
y, sr = sT.wav2waveform(audioFile)
```

```python
# the simplest box filter

def myBandFilterFilter(y, Fs, lb, ub):
    Y = np.fft.rfft(y)
    x = np.linspace(0, Fs/2, len(Y))
    Y[ x < lb ]=0
    Y[ x > ub ]=0
    return np.fft.irfft(Y)

y_fil = myBandFilterFilter(y, sr, 1000, ub=sr/2)

fig, ax = plt.subplots(2,3, figsize=(16,4))
ax[0,0].plot(y)
ax[1,0].plot(y_fil)
### spectrum
ax[0,1].plot(np.fft.rfft(y))
ax[1,1].plot(np.fft.rfft(y_fil))
### spectrogram
ax[0,2].specgram(y, Fs=sr)#, origin='bottom', aspect='auto')
ax[1,2].specgram(y_fil, Fs=sr)#10. * np.log10(A_filt), origin='bottom', aspect='auto')
print(":)")
```

## Butter_bandpass_filter

```python
lb = 500; hb = 24000; order = 2
y_filt = sT.butter_bandpass_filter(y, lb, hb, sr, order)
fig, ax = plt.subplots(2,2, figsize=(10,5))
ax[0,0].plot(y+1)
ax[0,1].plot(y_filt)
ax[1,0].specgram(y, Fs=sr)#, origin='bottom', aspect='auto')
ax[1,1].specgram(y_filt, Fs=sr)#10. * np.log10(A_filt), origin='bottom', aspect='auto')
print(":)")
```
