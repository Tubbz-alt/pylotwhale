#### filter 1: row/column normalization

```python
print np.shape(A)

def normalizeRowsOrColumns(M, ax = 1):
    norms_mat = np.max(np.abs(M), axis = ax)
    if (ax):
        nM = 1.0*M.T/norms_mat # divioisn column by column
        return nM.T
    else:
        nM = 1.0*M/norms_mat # divioisn column by column
        return nM
```

```python
ACnor = normalizeRowsOrColumns(A)
ARnor = normalizeRowsOrColumns(A,0)
Anor = normalizeRowsOrColumns(ARnor)
```
