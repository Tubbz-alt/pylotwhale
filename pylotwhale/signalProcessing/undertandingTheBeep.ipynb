### functions

```python
def time_line(wf_length, sampling_rate):
    """
    defines a the time vector from the sampling rate.
    aother alternative is to use
    np.linspace(start, stop, N)
    """
    
    vec = np.arange(0.0, 1.0*wf_length/sampling_rate, 1.0/sampling_rate)
    if(len(vec) == wf_length):
        return vec
    elif(len(vec) == wf_length+1): # longer time vec
        return np.delet(vec, len(vec)-1) # remve the last element
    elif(len(vec) == wf_length-1): # shorter time vec
        return np.append(vec, vec[-1]+1.0/sampling_rate)   
```

```python
def myHarmonic(f0, numHar, t, phase = 0.0):
    '''
    this function returns a oscilating dital signal (of t) with fundamental frequency f0, and numHar harmonics
    '''
    signal = 0.0
    for n in np.arange(1, numHar+1):
        signal = signal + np.sin(2.0*pi*f0*n*t + phase)
       # print n, numHar
    return signal/(numHar+1)

#pl.plot(myHarmonic(660,1, time_line(512*2, 192000)))
```

```python
def chop(sig):
    i = 0
    #    m = signal[i-1] - signal[i] 
    while True:
        m = sig[i+1] - sig[i] 
        i = i+1
        #print i, m
        if ( m >= 0):
            break
    newSig = sig[i-1:]
    x = np.arange( len(sig)-len(newSig),len(sig))
    
    return x,newSig

vec=np.array([10,9,10,8,7,1,2,3,4,5,10,9,8,7,9])
pl.plot(vec)
chopvec = chop(vec)
#print np.shape(chopvec), chopvec[0], chopvec[1]
pl.plot(chopvec[0], vec[chopvec[0]], 'r')
```
