```python
# Except for the first line, hashmark means comment.  the first line allows
# you to make the file an executable instead of calling it with python.

# make sure you have the following python packages installed.  sys and struct
# will come with any installation but the others you will likely have to 
# install yourself.

import sys
from scipy import *
import pylab as pl
import numpy as np
import scikits.audiolab as audiolab
import struct
```

```python
def show_Specgram(speech):
    '''
    Reads .wav file from STDIN and plots the spectrogram
    '''
    sound = audiolab.sndfile(speech,'read')
    # Reads wav file with audiolab
    sound_info = sound.read_frames(sound.get_nframes())
    # Extracts feature info from sound file with scipy module
    # sound_info will have 1 column for monaural data, 2 for stereo etc.
    figure( )
    subplots_adjust(hspace=.5)
    # get the sample rate, this will ensure the proper axes in the spectorgram.
    Fs = sound.get_samplerate()
    window1 = mlab.window_hanning
    window2 = np.hanning(256)
    nowindow = window_none
    #Generates soectrogram with matplotlib specgram

    print "sound info", sound_info
    
    #check the size of the array
    
    figure(figsize=(10,4))
    subplot(131)
    spectrogram = specgram(np.atleast_2d(sound_info)[0,:],Fs=Fs,window=nowindow,noverlap=100)
    spectrogram=specgram
    title('Spectrogram:  nowindow')
    subplot(132)
    spectrogram = specgram(np.atleast_2d(sound_info)[0,:],Fs=Fs,window=window1,noverlap=100)
    spectrogram=specgram
    title('Spectrogram:  Hanning')
    subplot(133)
    spectrogram = specgram(np.atleast_2d(sound_info)[0,:],Fs=Fs,window=window1,noverlap=100)
    spectrogram=specgram
    title('Spectrogram:  Hamming, overlap 100')
    show()
    sound.close()
    return spectrogram


#file = sys.argv[1]

#show_Specgram(file)
```
