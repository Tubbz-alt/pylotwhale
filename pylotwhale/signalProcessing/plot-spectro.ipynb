### wav plot

```python
tf = 1.0*(len(waveform)-0.1)/sRate
ff = sRate/2.0
time = np.arange(0.0, tf, 1.0/sRate)
print "Ntime",len(time),"Nwaveform", len(waveform),"final time", tf, "final freq", ff, "sampling rate", sRate

pl.figure(figsize=(10,3))
pl.subplot(121)
pl.title('waveform')
pl.plot(time[:], waveform[:])
pl.xlabel('time [s]')
pl.ylabel('amplitud')
pl.subplot(122)
pl.title('zoom')
pl.plot(time[:], waveform[:])
pl.xlabel('time [s]')
pl.ylabel('amplitud')
pl.xlim(0,0.001)
#pl.ylim(0,0.02)
#pl.savefig(baseN+'-waveForm.png') 
```

### spectrogram preview

```python
# setings
power = 10

N = 2**power
overp = 0.9
over = int(N*overp)
winN = 'hanning'#'blackman'
win =  sig.get_window(winN, N)#np.hamming(N)
Sr = sRate # just for the units

# spectro
#A = pl.specgram(waveform, Fs =Sr, NFFT = N, noverlap = over, window = win, scale_by_freq='log')[0]
A = pl.specgram(waveform, Fs =Sr, NFFT = N, noverlap = over, window = win)[0]
#pl.ylim(0,2000)

print "spec shapes:", np.shape(A)#, np.shape(spec1[0]), np.shape(spec2[0])

baseSpecN = baseN+'-%.2f%s%d'%(overp, winN, power)
print "in:", fileN,"\nout base name:", baseN
print "\nout spec base name:", baseSpecN
```
